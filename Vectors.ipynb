# Vectors

```
import numpy as np
import matplotlib.pyplot as plt
```

# Vector Plotting

```
plt.quiver(0,0,3,4)
plt.show
```

```
plt.quiver(0,0,3,4, scale_units='xy', angles='xy', scale=1)
plt.show
```

```
plt.quiver(0,0,3,4, scale_units='xy', angles='xy', scale=1)
plt.xlim(-10,10)
plt.ylim(-10,10)
plt.show
```

```
plt.quiver(0,0,3,4, scale_units='xy', angles='xy', scale=1)
plt.quiver(0,0,-3,4, scale_units='xy', angles='xy', scale=1)
plt.xlim(-10,10)
plt.ylim(-10,10)
plt.show
```

```
plt.quiver(0,0,3,4, scale_units='xy', angles='xy', scale=1, color='r')
plt.quiver(0,0,-3,4, scale_units='xy', angles='xy', scale=1, color='g')
plt.xlim(-10,10)
plt.ylim(-10,10)
plt.show
```

```
# function for dynamic vector plots
def vectors_func(vecs):
  colors=['r','b','g','y']
  i=0
  for vec in vecs:
    plt.quiver(vec[0], vec[1], vec[2], vec[3], scale_units='xy', angles='xy', scale=1, color=colors[i%len(colors)])
    print(colors[i%len(colors)], i)
    i+=1
  plt.xlim(-10,10)
  plt.ylim(-10,10)
  plt.show()
```

```
vectors_func([(0,0,3,5), (0,0,-3,5), (0,0,-4,-4),(0,0,8,-5)])
```

# Vector addition & subtraction

```
vector = [(0,0,3,5), (0,0,-3,5), (0,0,-4,-4),(0,0,8,-5)]
```

```
vectors_func([vector[0],vector[3]])
```

```
vector = [np.asarray([0,0,3,5]), np.asarray([0,0,-3,5]), np.asarray([0,0,-4,-4]), np.asarray([0,0,8,-5])]
```

```
vector[0]+vector[3]
```

```
vectors_func([vector[0],vector[1],vector[0]+vector[1]])
```

```
vectors_func([vector[0],vector[1],vector[0]-vector[1]])
```

# Vector dot product

```
vecs = [np.asarray([0,0,3,5]), np.asarray([0,0,-3,5]), np.asarray([0,0,-4,-4]), np.asarray([0,0,8,-5])]
```

```
vectors_func(vecs)
```

```
a = np.asarray([3,4])
b = np.asarray([4,-1])
```

```
a_dot_b = np.dot(a,b)
print(a_dot_b)
```

```
# I am trying to get the projection of a onto b
a_b = np.dot(a,b)/np.linalg.norm(b)
# a_b = a_dot_b/np.liaalg.norm(b)
print(a_b)
```

```
# Getting the vector of the projection of a onto b
#vec_a_b = (a_b/np.linalg.norm(b))*b
 vec_a_b = ((np.dot(a,b)/np.linalg.norm(b))/np.linalg.norm(b))*b
print(vec_a_b)
```
